# MathplotLib Tutorial : Bar chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

Simple bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))
print(xpos)
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue, width=0.2, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.2,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

<h3>Horizontal bar chart using barh function<h3>

```python
plt.barh(xpos-0.2,revenue, label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

# Matplotlib Tutorial: Histogram

in histograms, x axis contains a variable and y axis will be a frequency of that variable

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.95) # by default number of bins is set to 10
```

```python
#define the bin parameter 
plt.hist(blood_sugar,rwidth=0.5,bins=7)
```

```python
#define the bin parameter 
plt.hist(blood_sugar,rwidth=0.5,bins=10,color = 'Green')
```

```python
#define the bin parameter 
plt.hist(blood_sugar,rwidth=0.5,bins=7,color = 'Red')
```

Histogram showing normal, pre-diabetic and diabetic patients distribution
1.80-100: Normal
2.100-125: Pre-diabetic
3.80-100: Diabetic

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='r')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.80, color=['green','orange'],label=['men','women'])
plt.legend() #Multiple Data Samples
```

```python
#plot Step histrogram plot
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.80,histtype='step')
```

```python
#Horizontal Oriantation
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95,color = 'Red', orientation='horizontal',label = 'BloodSugar')
plt.legend()
```

# Matplotlib Tutorial: Pie Charts

Draw pie chart to track down home expenses

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

#Draw A pi char As perfect circle

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

# Show percentages for every pie. Specify radius to increase chart size

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True,autopct='%1.1f%%',radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0.1,0,0.1,0.2])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0.1,0,0.1,0.2],counterclock=True, startangle=90)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0],startangle= 90)
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=1, transparent=True)
```
